## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
#importing pandas, matplotlib, mounting google drive
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive

# reading the file and converting to dataframe
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head
#filtering dataframe to create new filtered dataframe views
netflix_subset = netflix_df[netflix_df['type']=='Movie'] #dataframe with only Movie in type column
netflix_movies = netflix_subset[["title","country","genre","release_year","duration"]] #dataframe with only "title","country","genre","release_year","duration" columns
short_movies = netflix_movies[netflix_movies.duration < 60] #dataframe with shorter than 60 mins movies

#associating every genre in dataframe with a color
colors = []
for label, row in netflix_movies.iterrows() :
    if row["genre"] == "Children" :
        colors.append("yellow")
    elif row["genre"] == "Documentaries" :
        colors.append("black")
    elif row["genre"] == "Stand-Up":
        colors.append("red")
    else:
        colors.append("white")

#creating a plot
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors) #plot of duration vs. release_year
#plot titles and axis labels
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

#show plot
plt.show()

#Answering the question: "Are we certain that movies are getting shorter?"
answer = "no"
```
